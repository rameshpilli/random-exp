```python
import newspaper
url = "https://www.latimes.com/politics/story/2020-10-24/trump-and-biden-barnstorm-battleground-states-as-race-moves-to-overdrive"

my_article = Article(url,language ='en')
my_article.download()
# print(my_article.html)
```

```python
my_article.parse()

#Extract the title 
print("Title..:", my_article.title)

#Extract the authors
print("Author..:", my_article.authors)

#Publishing date
print("Published on..:", my_article.publish_date)
```

```python
#Nlp on the article 
my_article.nlp()
#Extract summary
print("Summary...:", my_article.summary)

#Extract keywords
print("keywords..:", my_article.keywords)
```
