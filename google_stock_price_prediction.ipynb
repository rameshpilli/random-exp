```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range=(0,1))
training_set_scaled = sc.fit_transform(training_set)
```

```python
#Create a data structure with 60 timesteps and 1 output. 
#---> take data from day 1 - 60 and make predictions on the 61day

x_train = []
y_train = []
for i in range(60, len(training_set_scaled)):
    x_train.append(training_set_scaled[i-60:i,0])
    y_train.append(training_set_scaled[i,0])
# x_train
x_train,y_train = np.arange()
```

```python
import numpy as np
```

```python
x_train = np.array(x_train)
y_train = np.array(y_train)
```

```python
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1],1))
```

```python
#feature extraction -- wil choose featres that are to bee fed to the neural network 
#RNN
from keras.models import Sequential #linear stack of layers though which you can create a sequential model by passing the list though it .
from keras.layers import Dense #Regular deeply connected neural network layer---> used to chnaged the dimentions of the output vector 
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
#initialize the RNN
#fot the timeseries model like this---> basically use the regression 
regressor = Sequential()
```

```python
regressor.add(LSTM(units =50, return_sequences=True, input_shape = (x_train.shape[1],1)))
regressor.add(Dropout(0.2)) #for reducing over fitting in neural network 

regressor.add(LSTM(units=50, return_sequences=True))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units=50, return_sequences=True))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 50))
regressor.add(Dropout(0.2))

#need only one output .. adding the output layer. 
regressor.add(Dense(units=1))
```

```python
#Compile your RNN using an optimizer 
regressor.compile(optimizer='adam', loss = 'mean_squared_error')

#fitting the RNN to the training set
regressor.fit(x_train, y_train, epochs=100, batch_size=32)
```
