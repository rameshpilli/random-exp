```
import time
# import pprint
import pandas as pd
import configparser
from alpha_vantage.timeseries import TimeSeries

# api_key = "XXXXXXXXXXXXXX"
```

```
def get_api_key():
  config=configparser.ConfigParser()
  config.read("/content/drive/My Drive/Colab Notebooks/001_learning/config.ini")
  return config['openstockmap']['api']
```

```
timeseries = TimeSeries(key=get_api_key(), output_format='pandas')
data, metadata = timeseries.get_intraday(symbol='TSLA',interval='1min', outputsize='full') #min by min data
print(data)
# pd.DataFrame(dataa)
```

```
path = "/content/drive/My Drive/Colab Notebooks/csvs"
i = 1
while i==1:
  data, metadata = timeseries.get_intraday(symbol='TSLA',interval='1min', outputsize='full')
  data.to_csv(path+'tsla_stock_min.xlsx')
  time.sleep(60)
```

```
#day trading is done by using volatility of the stock
close_data = data['4. close']
#calculate the pct change for closing values in real time
pct = close_data.pct_change()
```

```
#volatile stocks we only care about the last value 
last_change_value = pct[-1]

#stock alert and volatility
if abs(last_change_value) > 0.0004:
  print("Tsla alert:", last_change_value)
```
